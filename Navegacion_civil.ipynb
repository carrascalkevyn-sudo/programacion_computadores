# INTRODUCCIÓN

Crear una **Tabla de Contenido (TOC)** en Google Colab es fundamental para mantener el orden, especialmente cuando tus cuadernos (notebooks) se vuelven extensos. A diferencia de un documento de texto tradicional, en Colab tienes dos formas principales de gestionar la navegación: una automática y una personalizada.

---

## 1. El Panel de Navegación Automático

Google Colab genera automáticamente una tabla de contenido basada en las **celdas de texto** que utilizan encabezados de Markdown.

* **Cómo activarlo:** Haz clic en el icono de lista (tres líneas horizontales) en la barra lateral izquierda.
* **Cómo funciona:** Si escribes `# Título 1` o `## Subtítulo`, aparecerán instantáneamente en ese panel, permitiéndote saltar de una sección a otra con un solo clic.

---

## 2. Tabla de Contenido Manual (Dentro del Notebook)

Si prefieres tener un índice al principio de tu cuaderno para que otros usuarios lo vean fácilmente, puedes crear **enlaces internos** usando Markdown. La sintaxis es la siguiente:

1. **Crea el encabezado:** Define tu sección con `# Mi Sección`.
2. **Crea el enlace:** En una celda de texto aparte, usa el formato `[Nombre del Enlace](#nombre-de-la-seccion-con-guiones)`.

> **Nota importante:** Colab convierte automáticamente los títulos a minúsculas y reemplaza los espacios por guiones para generar el identificador del enlace.

### Ejemplo de sintaxis:

```markdown
# Índice
1. [Introducción](#introduccion)
2. [Análisis de Datos](#analisis-de-datos)
3. [Conclusiones](#conclusiones)

---
## Introducción
Contenido de la intro...

```

---

## Beneficios de una buena estructura

* **Legibilidad:** Facilita que otros entiendan el flujo de tu código.
* **Productividad:** No pierdes tiempo haciendo scroll infinito buscando una función específica.
* **Interactividad:** Convierte tu cuaderno en un reporte técnico profesional.

# FUNDAMENTO TEÓRICO

El fundamento técnico para crear una tabla de contenido en Google Colab (y en los cuadernos de Jupyter en general) se basa en el lenguaje de marcado **Markdown** y en la interpretación que hace el navegador del **HTML** generado.

Aquí te explico los pilares teóricos que hacen posible esta navegación:

---

## 1. El Lenguaje Markdown como Estructura

Markdown es un lenguaje de marcado ligero que permite aplicar formato a texto plano. En el contexto de un notebook, las celdas de texto actúan como la capa de presentación. La jerarquía se establece mediante el símbolo `#`:

* **Estructura Jerárquica:** Los encabezados (`#`, `##`, `###`) no solo cambian el tamaño de la fuente, sino que definen la **arquitectura de la información** del documento. Google Colab escanea estos niveles para construir el árbol de navegación lateral.

## 2. Anclas e Identificadores (ID)

El "secreto" de la navegación interactiva reside en las **anclas HTML**. Cuando Colab renderiza una celda de texto con un encabezado, asigna automáticamente un identificador único al elemento.

* **Mecanismo de Enlace:** Un enlace interno funciona mediante la referencia a un ID específico dentro de la misma página web.
* **Normalización de IDs:** Colab transforma los títulos en identificadores válidos para URLs. Por ejemplo, el título `## Procesamiento de Datos` se convierte internamente en el ancla `#procesamiento-de-datos`.

---

## 3. Hipervínculos Internos (Fragment Identifiers)

La navegación se completa utilizando lo que en desarrollo web se conoce como *Fragment Identifiers*. Al hacer clic en un elemento del índice, el navegador busca el elemento cuya propiedad `id` coincida con el fragmento después del símbolo `#` en la URL y desplaza la vista (scroll) hacia esa posición.

### Resumen del flujo lógico:

| Componente | Función Teórica |
| --- | --- |
| **Marcado (#)** | Define la jerarquía y crea el punto de destino (destino del salto). |
| **Sintaxis `[Texto](#ancla)**` | Crea el disparador (el botón o link) que solicita el salto. |
| **Renderizado** | El entorno Colab traduce el Markdown a HTML, vinculando ambos puntos. |

---

## 4. El Panel de Control (Sidebar)

A diferencia de los enlaces manuales, el panel izquierdo de Colab utiliza un **script dinámico**. Este lee constantemente el DOM (Document Object Model) del cuaderno y genera una interfaz de usuario (UI) que actúa como un "mapa" persistente, permitiendo que la navegación sea independiente de la celda en la que te encuentres.

# Fórmula de Ramanujan para pi

Srinivasa Ramanujan descubrió varias series notables para calcular π. Una de las más conocidas y utilizadas, que converge muy rápidamente, es la siguiente:

$$\frac{1}{\pi} = \frac{2\sqrt{2}}{9801} \sum_{n=0}^{\infty} \frac{(4n)!}{(n!)^4} \frac{26390n + 1103}{396^{4n}}$$

Esta fórmula es notable por su increíble velocidad de convergencia, lo que significa que con solo unos pocos términos se pueden obtener muchas cifras decimales de pi.
